## Declination of the Sun
$\delta_{\odot}=-\arcsin\left[\sin\epsilon \cos \left(\dfrac{360^{\circ}}{365.24}(N+10)\right)\right]$

```python
N = 171
omega = 2*pi / 365.24
ecl = radians(23.44)

delta = -asin(sin(ecl) * cos(omega * (N + 10)))
print('declination = {:.2f} deg'.format(degrees(delta)))
```

```python
import astropy.units as u
from astropy.coordinates import \
SkyCoord, EarthLocation, AltAz, get_sun

obs = EarthLocation(lat = 53 * u.deg + 28 * u.arcmin + 49 * u.arcsec,
                    lon = 10 * u.deg+14*u.arcmin + 23 * u.arcsec)
phi = obs.lat
```

```python
elapsed_night =elapsed[np.where(sun_local.alt<0)]
betelgeuse_night = betelgeuse_local.alt[np.where(sun_local.alt<0)]
plt.plot(elapsed.to(u.h), sun_local.alt, color='orange', label='Sun')
plt.plot(elapsed.to(u.h), betelgeuse_local.alt, 
        color='red', linestyle=':', label='Betelgeuse (daylight)')
plt.plot(elapsed_night.to(u.h), betelgeuse_night, color='red', label='Betelgeuse (night)')
plt.xlabel("Time from noon [h]")
plt.xlim(0, 24)
plt.xticks(np.arange(13) *2)
plt.ylim(0, 60)
plt.ylabel('Altitude [deg]')
plt.legend(loc='upper right')
plt.savefig('Betelgeuse_obs_window.pdf')
```

## 2.3
$\delta_{\odot}=-\arcsin\left[\sin\epsilon_0 \cos\left(\dfrac{360^{\circ}}{365.24}(N+10) +e\dfrac{360^{\circ}}{\pi}\sin\left[\dfrac{360^{\circ}}{365.24}(N-2)\right]\right)\right]$

where $e=0.0167$

```python
e = 0.0167
N = np.array(N)
delta1 = np.rad2deg(-np.arcsin(np.sin(ecl) * np.cos(omega * (N+10))))
delta2 = np.rad2deg(-np.arcsin(np.sin(ecl)*np.cos(omega* (N + 10) + e * 2 *\
                                        np.sin(omega * (N -2 )))))
print('circular [deg]   ecliptical [deg]  relative deviation[%]')
for row in zip(delta1, delta2, (delta2 - delta1)/delta2 * 100):
    print(f"{row[0]:2.3f}               {row[1]:2.3f}               {row[2]:3.2f}")
```

```python
#2.6
N = np.arange(365)
N = np.arange(365)
omega =  2* math.pi/365.24
ecl = math.radians(23.44)

delta1 = -np.arcsin(np.sin(ecl)*np.cos(omega* (N + 10) + e * 2 *\
                                        np.sin(omega * (N -2 ))))
h1 = np.arccos(-np.tan(delta1) * math.tan(phi.radian))
T1 = (np.degrees(2 * h1)/360) * u.sday.to(u.h)

delta2 = -np.arcsin(np.sin(ecl)*np.cos(omega* (N + 10)))
h2 = np.arccos(-np.tan(delta2) * math.tan(phi.radian))
T2 = (np.degrees(2 * h2)/360) * u.sday.to(u.h)


T3 = []
for N_ in range(365):
    elapsed = np.arange(0 , 24 * 60, 1) * u.min
    time = Time('2020-1-1 00:00:00') - 8*u.h + N_*u.day + elapsed
    sun = get_sun(time)
    frame_local_24h = AltAz(obstime=time, location=obs)
    sun_local = sun.transform_to(frame_local_24h)
    daytime =elapsed[np.where(sun_local.alt>0)]
    day_length = daytime[-1] - daytime[0]
    T3.append(day_length.to(u.h).value)
```

```python
plt.plot(N, T1, label='corrected day length')
plt.plot(N, T2, label='no correction')
plt.plot(N, T3, label='actual day length')
plt.xlabel("Day")
plt.xlim(0, 364)
plt.ylabel('Day length [hr]')
plt.ylim(9, 15)
plt.legend(loc='upper right');
plt.savefig("daylength_WHU.pdf")
```

```python
#2.7
import astropy.units as u
from astropy.coordinates import \
SkyCoord, EarthLocation, AltAz, get_sun

obs = EarthLocation(lat = 53 * u.deg + 28 * u.arcmin + 49 * u.arcsec,
                    lon = 10 * u.deg+14*u.arcmin + 23 * u.arcsec)
phi = obs.lat
from astropy.time import Time
utc_shift = 1 * u.hour
noon_cest = Time('2020-12-31 12:00:00') - utc_shift
elapsed = np.arange(0 , 24 * 60, 5) * u.min
time = noon_cest + elapsed
frame_local_24h = AltAz(obstime=time, location=obs)
star_list = ['Betelgeuse', 'Polaris', 'Sirius A']

sun = get_sun(time)
sun_local = sun.transform_to(frame_local_24h)
elapsed_night =elapsed[np.where(sun_local.alt<-18*u.deg)]
plt.plot(elapsed.to(u.h), sun_local.alt, color='orange', label='Sun')
for star in star_list:
    star_pos = SkyCoord.from_name(star)
    star_local = star_pos.transform_to(frame_local_24h)
    star_night = star_local.alt[np.where(sun_local.alt<-18 * u.deg)]
    plt.plot(elapsed.to(u.h), star_local.alt, linestyle=':', label=f'{star} (daylight)')
    plt.plot(elapsed_night.to(u.h), star_night, label=f'{star} ( astronomical night)')
plt.xlabel("Time from noon [h]")
plt.xlim(0, 24)
plt.xticks(np.arange(13) *2)
plt.ylim(0, 60)
plt.ylabel('Altitude [deg]')
plt.legend(loc='upper right',fontsize=8);
plt.savefig('Exercise 2.7.pdf')
```
